### What about Raw Data
We need to read hour data sets

```python
import pandas as pd
path = 'ag_news_csv/train.csv'
df = pd.read_csv(path, sep=',', names = ['class', 'title', 'body'])
```

```python
df
```

```python
df['body'][0].split()
```

```python
df['body'].shape
```

```python
list_texts = [ text.split() for text in df['body'] ]
```

```python
list_texts
```

```python
import collections
allterms = []
for ls in list_texts:
    allterms += ls
counter = collections.Counter(allterms)
```

```python
counter['holes.']
```

```python
diffterms = list(set(allterms))
```

```python
diffterms
```

```python
hashterms = {}
for i, word in enumerate(diffterms):
    hashterms.update({word:i})
```

```python
hashterms
```

```python
irow = []
jcol = []
data = []
for j, text in enumerate(list_texts):
    for word in text:
        irow.append(hashterms[word])
        jcol.append(j)
        data.append(float(counter[word]))
```

```python
print(len(data))
print(len(list_texts)*len(hashterms))
```

```python
M = len(hashterms)
N = len(list_texts)
```

```python
N
```

```python
M
```

```python
from scipy.sparse import coo_matrix 
compress_matrix = coo_matrix((data, (irow, jcol)), shape=(M, N))
```

```python
del data
del irow
del jcol
del list_texts
del counter
del diffterms 
del hashterms
```

```python
compress_matrix.data.nbytes
```

```python
CA = compress_matrix.tocsc()
```
